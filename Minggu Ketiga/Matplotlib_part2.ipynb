```python
# import pandas
import pandas as pd
```

```python
data = pd.read_csv('data_bar (2).csv')
```

```python
data.head()
```

#### Counter

```python
# import modul Counter
from collections import Counter
```

```python
# Counter digunakan untuk menghitung jumlah anggota suatu set/list
c = Counter({'python', 'js', 'c', 'c++'})
c
```

```python
c.update(['c++', 'python'])
c
```

```python
d = Counter(['python', 'js', 'c', 'c++'])
d
```

#### Apply it to our data

```python
ids = data['Responder_id']
lang = data['LanguagesWorkedWith']

lang_counter = Counter()
```

```python
for response in lang:
    lang_counter.update(response.split(';'))
```

```python
print(lang_counter)
```

```python
# Menampilkan 15 data teratas
print(lang_counter.most_common(15))
```

```python
language = []
popularity = []

for item in lang_counter.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
```

```python
print(language)
print(popularity)
```

#### Add to Bar Chart

```python
plt.bar(language, popularity)

plt.title('Most Popular Language')
plt.xlabel('Language')
plt.ylabel('Number of Users')

plt.show()
```

Dapat dilihat bahwa data di label x tidak terlihat, sehingga perlu diubah menjadi horizontal

#### Horizontal Bar

```python
# Membuat bar hoeizontal
plt.barh(language, popularity)

plt.title('Most Popular Language')
# plt.ylabel('Language')
plt.xlabel('Number of Users')

plt.show()
```

```python
# Membalik Urutan Data -> Gunakan method reverse
language.reverse()
popularity.reverse()

plt.barh(language, popularity)

plt.title('Most Popular Language')
# plt.ylabel('Language')
plt.xlabel('Number of Users')

plt.show()
```

## Subplot

```python
data2 = pd.read_csv('data_sub (2).csv')
data2.head()
```

```python
# Pisahkan data
ages = data2['Age']
dev_salaries = data2['All_Devs']
py_salaries = data2['Python']
js_salaries = data2['JavaScript']
```

#### Mendefinisikan Subplot

```python
# Gunakan style 'seaborn'
plt.style.use('seaborn')

# Mendefinisikan subplot
fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots()
ax.plot(ages, py_salaries, label='Python')
ax.plot(ages, js_salaries, label='JavaScript')
ax.plot(ages, dev_salaries, label='All Devs')

ax.legend()
ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages');
ax.set_ylabel('Median Salary (USD)');

plt.show()
```

#### Subplot lebih dari satu

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1)
ax1.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')
ax2.plot(ages, dev_salaries, label='All Devs')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)');

ax2.legend()
ax2.set_xlabel('Ages');
ax2.set_ylabel('Median Salary (USD)');

plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')
ax2.plot(ages, dev_salaries, label='All Devs')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages');
ax1.set_ylabel('Median Salary (USD)');

ax2.legend()
ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages');

plt.show()
```

#### Figure lebih dari satu

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
ax1.plot(ages, py_salaries, label='Python', linestyle='--')
ax2.plot(ages, js_salaries, label='JavaScript')
ax2.plot(ages, dev_salaries, label='All Devs')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages');
ax1.set_ylabel('Median Salary (USD)');

ax2.legend()
ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages');
ax2.set_ylabel('Median Salary (USD)');

plt.show()
```
